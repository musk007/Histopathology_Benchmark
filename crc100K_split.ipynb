```python
import os
import pandas as pd
norm_images = "/home/roba/benchmark_req/data/Kather_data/NCT-CRC-HE-100K"
NoNorm_images = "/home/roba/benchmark_req/data/CRC_100K/NCT-CRC-HE-100K-NONORM"
```

```python
import random

def pick_100(names, seed=None):
    # if seed is not None:
    #     random.seed(seed)
    k = min(3000, len(names))
    return random.sample(names, k)  # unique picks

norm_cls_0 = os.listdir(os.path.join(norm_images, "ADI"))
norm_cls_1 = os.listdir(os.path.join(norm_images, "DEB"))
norm_cls_2 = os.listdir(os.path.join(norm_images, "LYM"))
norm_cls_3 = os.listdir(os.path.join(norm_images, "MUC"))
norm_cls_4 = os.listdir(os.path.join(norm_images, "MUS"))
norm_cls_5 = os.listdir(os.path.join(norm_images, "NORM"))
norm_cls_6 = os.listdir(os.path.join(norm_images, "STR"))
norm_cls_7 = os.listdir(os.path.join(norm_images, "TUM"))
norm_cls_8 = os.listdir(os.path.join(norm_images, "BACK"))

NOnorm_cls_0 = os.listdir(os.path.join(NoNorm_images, "ADI"))
NOnorm_cls_1 = os.listdir(os.path.join(NoNorm_images, "DEB"))
NOnorm_cls_2 = os.listdir(os.path.join(NoNorm_images, "LYM"))
NOnorm_cls_3 = os.listdir(os.path.join(NoNorm_images, "MUC"))
NOnorm_cls_4 = os.listdir(os.path.join(NoNorm_images, "MUS"))
NOnorm_cls_5 = os.listdir(os.path.join(NoNorm_images, "NORM"))
NOnorm_cls_6 = os.listdir(os.path.join(NoNorm_images, "STR"))
NOnorm_cls_7 = os.listdir(os.path.join(NoNorm_images, "TUM"))
NOnorm_cls_8 = os.listdir(os.path.join(NoNorm_images, "BACK"))

norm_lists   = [norm_cls_0,norm_cls_1,norm_cls_2,norm_cls_3,norm_cls_4,norm_cls_5,norm_cls_6,norm_cls_7,norm_cls_8]
nonorm_lists = [NOnorm_cls_0,NOnorm_cls_1,NOnorm_cls_2,NOnorm_cls_3,NOnorm_cls_4,NOnorm_cls_5,NOnorm_cls_6,NOnorm_cls_7,NOnorm_cls_8]
```

```python
picked_norm = []
picked_NoNorm = []
for lst in norm_lists:
    picked = pick_100(lst)
    picked_norm.extend(picked)

for lst in nonorm_lists:
    picked = pick_100(lst)
    picked_NoNorm.extend(picked)
```

```python

norm_df = pd.DataFrame(columns=["image","label","label_text","text_style_4"])
# for im in picked_NoNorm:
for im in picked_norm:
    folder = im.split("-")[0]
    if folder == "ADI":
        label = 0
        label_text = "adipose tissue"
    elif folder == "DEB":
        label = 1
        label_text = "debris"
    elif folder == "LYM":
        label = 2
        label_text = "lymphocytes"
    elif folder == "MUC":
        label = 3
        label_text = "mucus"
    elif folder == "MUS":
        label = 4
        label_text = "smooth muscle"
    elif folder == "NORM":
        label = 5
        label_text = "normal colon mucosa"
    elif folder == "STR":
        label = 6
        label_text = "cancer-associated stroma"
    elif folder == "TUM":
        label = 7
        label_text = "colorectal adenocarcinoma epithelium"
    elif folder == "BACK":
        label = 8
        label_text = "background"
    df = pd.DataFrame({
        "image":[os.path.join(norm_images,folder, im)],
        # "image":[os.path.join(NoNorm_images,folder, im)],
        "label":label,
        "label_text":label_text,
        "text_style_4":f"An H&E image patch of {label_text} tissue."
        })
    norm_df = pd.concat([norm_df, df], ignore_index=True)
```

```python
# norm_df.to_csv("/benchmark_req/data/text_files/Kather_NoNorm_test.csv", index=False)
norm_df.to_csv("/benchmark_req/data/text_files/Kather_norm_test.csv", index=False)
```
